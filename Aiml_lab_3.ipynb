```python
def sum_odd_even_cubes(num_set):
    sum_odd_cube = 0
    sum_even_cube = 0

    for num in num_set:
        if num % 2 == 0:
            sum_even_cube += num ** 3
        else:
            sum_odd_cube += num ** 3

    return sum_odd_cube, sum_even_cube

numbers = {1, 2, 3, 4, 5}
odd_cube_sum, even_cube_sum = sum_odd_even_cubes(numbers)

print("SumOddCube:", odd_cube_sum)
print("SumEvenCube:", even_cube_sum)
```

```python

def matrix_operations(matrix):

    rows = len(matrix)
    cols = len(matrix[0])


    transpose = []
    for j in range(cols):
        new_row = []
        for i in range(rows):
            new_row.append(matrix[i][j])
        transpose.append(new_row)


    row_sum = []
    for row in matrix:
        total = 0
        for num in row:
            total = total + num
        row_sum.append(total)


    col_sum = []
    for j in range(cols):
        total = 0
        for i in range(rows):
            total = total + matrix[i][j]
        col_sum.append(total)

    inverse = "Inverse not possible"

    if rows == 2 and cols == 2:
        a = matrix[0][0]
        b = matrix[0][1]
        c = matrix[1][0]
        d = matrix[1][1]

        det = (a * d) - (b * c)

        if det != 0:
            inverse = [
                [ d / det, -b / det],
                [-c / det,  a / det]
            ]


    return transpose, row_sum, col_sum, inverse




matrix = [
    [1, 2],
    [3, 4]
]

transpose, row_sum, col_sum, inverse = matrix_operations(matrix)


print("Original Matrix:", matrix)
print("Transpose:", transpose)
print("Row-wise Sum:", row_sum)
print("Column-wise Sum:", col_sum)
print("Inverse:", inverse)
```

```python
import numpy as np


A = np.array([
    [1, -2, 3],
    [-1, 3, -1],
    [2, -5, 5]
])

B = np.array([9, -6, 17])

solution = np.linalg.solve(A, B)

x = solution[0]
y = solution[1]
z = solution[2]

print("Value of x:", x)
print("Value of y:", y)
print("Value of z:", z)
```

```python
import pandas as pd


url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"

columns = [
    'age', 'workclass', 'fnlwgt', 'education', 'education_num',
    'marital_status', 'occupation', 'relationship', 'ethnicity',
    'gender', 'capital_gain', 'capital_loss', 'hours_per_week',
    'country_of_origin', 'income'
]

df = pd.read_csv(url, header=None, names=columns, sep=', ', engine='python')

print(df.head(10))



print("Number of rows and columns:", df.shape)
print("\nData types:")
print(df.dtypes)
print("\nMemory usage:")
print(df.memory_usage())



print("\nFirst row:")
print(df.iloc[0])

print("\nFirst 10 rows:")
print(df.head(10))

print("\nRows 10 to 15:")
print(df.iloc[10:16])


print("\nLast two rows:")
print(df.tail(2))

print("\nOdd rows from columns 3 to 5:")
print(df.iloc[1::2, 2:5])
```
