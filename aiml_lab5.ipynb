```python
import pandas as pd


df = pd.read_csv("world_alcohol - world_alcohol.csv")

print("Original Dataset:")
display(df)
```

```python
print("\nExercise 1: Rows without NaN/empty values")
df_no_nan = df.dropna()
display(df_no_nan)
```

```python
print("\nExercise 2: Rows with NaN/empty values")
df_with_nan = df[df.isna().any(axis=1)]
display(df_with_nan)
```

```python
print("\nExercise 3: Dataset after deleting rows with NULL values")
df_deleted_null = df.dropna()
display(df_deleted_null)
```

```python
print("\nExercise 4: Replace NULL in 'Display Value' with MEAN")
mean_value = df["Display Value"].mean()
df_mean_filled = df.copy()
df_mean_filled["Display Value"].fillna(mean_value, inplace=True)
display(df_mean_filled)
```

```python
print("\nExercise 5: Replace NULL in 'Display Value' with 1.5")
df_fixed_value = df.copy()
df_fixed_value["Display Value"].fillna(1.5, inplace=True)
display(df_fixed_value)
```

```python
print("\nExercise 6: Every 5th row starting from index 1")
df_every_5th = df.iloc[1::5]
display(df_every_5th)
```

```python
print("\nExercise 7: Rename columns")

# Rename ALL columns
df_renamed_all = df.copy()
df_renamed_all.columns = ["Year", "Region", "Country", "Alcohol Type", "Avg Consumption"]
display(df_renamed_all)

# Rename ONLY SOME columns
df_renamed_some = df.rename(columns={
    "Display Value": "Avg Consumption",
    "Beverage Types": "Alcohol Type"
})
display(df_renamed_some)
```

```python
print("\nExercise 8: Rows with index ending in 0")
df_index_ending_0 = df[df.index % 10 == 0]
display(df_index_ending_0)
```

```python
# Exercise 9: Average consumption of wine per person greater than 2

wine_data = df[df["Beverage Types"] == "Wine"]
wine_data = wine_data.dropna(subset=["Display Value"])

wine_per_person_gt_2 = wine_data[wine_data["Display Value"] > 2]

display(wine_per_person_gt_2)
```
