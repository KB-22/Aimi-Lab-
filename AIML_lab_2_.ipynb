```python
# Question3 WAP to add all the numbers from 1 to n and n is given by user.
n = int(input("Enter a number: "))

sum = 0

for i in range(1, n + 1):
    sum = sum + i

print("Sum of numbers from 1 to", n, "is:", sum)
```

```python
# Question 4 WAP to find whether the given number is an Armstrong number.
num = int(input("Enter a number: "))

temp = num
sum = 0


digits = len(str(num))


while temp > 0:
    digit = temp % 10
    sum = sum + (digit ** digits)
    temp = temp // 10

if sum == num:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

```python
# Question 5 WAP to find average of first n numbers.
n = int(input("Enter a number: "))

sum = 0

for i in range(1, n + 1):
    sum = sum + i

average = sum / n

print("Average of first", n, "numbers is:", average)
```

```python
# Question 6 6. WAP to sum the series 1^2 / 1 + 2 ^2 / 2 + ……… + n ^2 / n
n = int(input("Enter the value of n: "))

sum = 0

for i in range(1, n + 1):
    sum = sum + (i**2) / i

print("Sum of the series is:", sum)
```

```python
# Question 8 WAP to print the sum of the series: 1 - x + x^2 /2! - x^3 / 3! + ……
x = float(input("Enter the value of x: "))
n = int(input("Enter number of terms: "))

sum = 0
fact = 1

for i in range(n):
    if i > 0:
        fact = fact * i

    term = (x**i) / fact
    if i % 2 == 1:
        term = -term

    sum = sum + term

print("Sum of the series is:", sum)
```
